## **1. Text Cleaning (Lowercase + Remove Punctuation)**

```python
# Import the libraries we need
import nltk            # NLTK = Natural Language Toolkit, the main NLP library in Python
import string          # gives us a ready-made list of punctuation characters (.,!? etc.)

# This is our sample input text (a "corpus" in NLP terms)
text = "Hello! My name is Shubham Sarwar and I am studying Natural Language Processing."

# STEP 1: Convert everything to lowercase.
# Why? Because "Hello" and "hello" are the SAME word but the computer treats them as different.
# Lowercasing makes the text uniform so counting & matching work correctly.
text_lower = text.lower()

# STEP 2: Remove punctuation.
# Why? Punctuation marks (! . ?) usually don't add meaning for tasks like counting words.
# str.maketrans('', '', string.punctuation) builds a "translation table" that maps
# every punctuation character to nothing (i.e., delete it). translate() applies that table.
text_clean = text_lower.translate(str.maketrans('', '', string.punctuation))

print("Original Text :", text)
print("\nLowercase Text:", text_lower)
print("\nAfter Removing Punctuation:", text_clean)
```

## **2. Tokenization**

Tokenization = splitting text into smaller pieces called **tokens**.  
- **Sentence tokenization** → splits paragraph into sentences.  
- **Word tokenization** → splits sentence into words.

It is the FIRST step of almost every NLP task.


```python
import nltk

# These are small data files NLTK needs to know HOW to split sentences/words.
# 'punkt' / 'punkt_tab' contain the trained sentence-splitter model.
nltk.download('punkt')
nltk.download('punkt_tab')

# Import the two tokenizer functions
from nltk.tokenize import sent_tokenize, word_tokenize

# Sample text containing two sentences
text = "Shubham Sarwar studies NLP. He practices Python programming every day."

# Split the text into SENTENCES (returns a list of sentences)
sent_tokens = sent_tokenize(text)

# Split the text into WORDS (returns a list of words; punctuation becomes its own token)
word_tokens = word_tokenize(text)

print("Sentence Tokens:")
print(sent_tokens)
print("\nWord Tokens:")
print(word_tokens)
```

## **3. Stop-word Removal**

**Stop-words** are very common words (is, the, in, and, a …) that occur in almost every sentence
and usually don't help the computer understand the *meaning*. We remove them to focus on the
words that actually carry information.


```python
import nltk
from nltk.corpus import stopwords          # NLTK ships with a list of stop-words for many languages
from nltk.tokenize import word_tokenize

# Download the stopwords list (only needed the first time)
nltk.download('stopwords')

text = "Shubham is studying NLP in the computer lab at college."

# 1. Tokenize the (lowercased) text into individual words
words = word_tokenize(text.lower())

# 2. Load the English stopword list. A set() is used because checking "word in set" is fast.
stop_words = set(stopwords.words("english"))

# 3. Keep only the words that are NOT stop-words.
#    filter() iterates over `words` and keeps each word for which the lambda returns True.
filtered_words = list(filter(lambda w: w not in stop_words, words))

print("Original Words :", words)
print("\nAfter Stopword Removal:", filtered_words)
```

## **4. Stemming**

**Stemming** chops off the ends of words to reduce them to a base/root form.  
- studying → studi  
- learning → learn  
- happily → happili  

It is FAST but the result is sometimes not a real dictionary word (e.g. "studi").  
We use the **Porter Stemmer**, the most popular stemming algorithm.


```python
from nltk.stem import PorterStemmer       # Porter's algorithm – classic stemming method
from nltk.tokenize import word_tokenize

text = "Shubham is studying and learning Natural Language Processing."

# Tokenize and lowercase
words = word_tokenize(text.lower())

# Create the stemmer object
stemmer = PorterStemmer()

# Apply .stem() to every word in the list using map()
stemmed_words = list(map(stemmer.stem, words))

print("Original Words:", words)
print("\nStemmed Words :", stemmed_words)
```

## **5. Lemmatization**

**Lemmatization** also reduces a word to its base form, BUT it uses a dictionary so the result
is always a real word (called the *lemma*).  
- studying → study (if we tell it the word is a verb)  
- techniques → technique  
- was → be

It is SLOWER than stemming but more accurate.


```python
from nltk.stem import WordNetLemmatizer   # uses the WordNet dictionary
from nltk.tokenize import word_tokenize
import nltk

# Download WordNet (a big English lexical database)
nltk.download('wordnet')

text = "Shubham was studying NLP concepts and learning new techniques."

words = word_tokenize(text.lower())

# Create the lemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization to each word.
# Note: by default the lemmatizer treats every word as a NOUN, that is why
# "studying"/"learning" are not reduced. To handle verbs we would pass pos='v'.
lemmatized_words = list(map(lemmatizer.lemmatize, words))

# Bonus: lemmatize verbs explicitly to see the difference
lemmatized_verbs = [lemmatizer.lemmatize(w, pos='v') for w in words]

print("Original Words            :", words)
print("\nLemmatized (default/noun) :", lemmatized_words)
print("\nLemmatized as verbs       :", lemmatized_verbs)
```

## **6. Word Frequency (Morpheme / Root Frequency Analysis)**

Counts how often each word/morpheme appears in the text. This helps us find the
**most important words** in a document — used in keyword extraction, summarization,
text classification, etc.


```python
from collections import Counter            # Counter quickly counts how many times each item appears
from nltk.tokenize import word_tokenize

text = "Shubham studies NLP. Shubham practices Python for NLP assignments."

# Tokenize + lowercase so "NLP" and "nlp" are counted together
words = word_tokenize(text.lower())

# Counter returns a dictionary-like object: {word: count}
freq = Counter(words)

print("Word Frequency:")
print(freq)

print("\nTop 3 most common words:")
print(freq.most_common(3))
```
