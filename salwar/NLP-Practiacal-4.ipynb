```python
import nltk
import re
import math
import random

nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import MLE
```

**Text Corpus**

```python
sentences = """
The morning after the upheaval of the night protests, the city was surreally quiet. Waking in the parking garage, Eleanor lifted herself out of the nest of old coats and her backpack on the floor of the backseat. As she drove away from the one sanctuary she could think of as streets were shut down, Eleanor saw evidence of the night's violence in the strewn litter, broken glass, and the watchful police presence.

She had always believed that the truth was a fixed point, but now it felt like a shadow at the edge of every moment. People in this town paid attention to everyone else's details: a car missing from a driveway, a skipped shift at work, or one less body tucked into the pew on Sunday morning. This was how they governed themselves—with the understanding that there was no such thing as a secret.

Eleanor turned the radio dial, but only static filled the air. She thought of the library back home, filled with books that smelled of old paper and dust. She wondered if her own story would ever be written, or if it would simply vanish, washed away by an alien sea of indifference.
"""
```

**Preprocessing**

```python
from nltk.tokenize import sent_tokenize

individual_sentences = sent_tokenize(sentences)

tokenized = [
    [word for word in word_tokenize(s.lower()) if word.isalpha()]
    for s in individual_sentences
]

print("\n" + "="*60)
print("STEP 2: TOKENIZATION OUTPUT")
print("="*60)

for i, t in enumerate(tokenized):
    print(f"Sentence {i+1}: {t}")

tokenized_with_tokens = [
    ['<s>'] + tokens + ['</s>']
    for tokens in tokenized
]
```

**Generate N-grams**

```python
all_unigrams, all_bigrams, all_trigrams = [], [], []

for tokens in tokenized_with_tokens:
    all_unigrams  += list(ngrams(tokens, 1))
    all_bigrams   += list(ngrams(tokens, 2))
    all_trigrams  += list(ngrams(tokens, 3))

print("\n" + "="*60)
print("STEP 3: N-GRAM GENERATION")
print("="*60)

print("Sample Unigrams:", all_unigrams[:10])
print("Sample Bigrams:", all_bigrams[:10])
print("Sample Trigrams:", all_trigrams[:10])
```

**Frequency Count**

```python
unigram_counts = FreqDist(all_unigrams)
bigram_counts  = FreqDist(all_bigrams)
trigram_counts = FreqDist(all_trigrams)

print("\n" + "="*60)
print("STEP 4: FREQUENCY COUNTS")
print("="*60)

print("\nTop Unigrams:")
for k, v in unigram_counts.most_common(10):
    print(k[0], ":", v)

print("\nTop Bigrams:")
for k, v in bigram_counts.most_common(10):
    print(k, ":", v)

print("\nTop Trigrams:")
for k, v in trigram_counts.most_common(10):
    print(k, ":", v)
```

**Probabilities**

```python
vocab = set([w[0] for w in unigram_counts])
V = len(vocab)

def bigram_prob(w1, w2, smoothing=False):
    count = bigram_counts.get((w1, w2), 0)
    base = unigram_counts.get((w1,), 0)

    if smoothing:
        return (count + 1) / (base + V)
    else:
        return count / base if base != 0 else 0

print("\n" + "="*60)
print("STEP 5: PROBABILITY CALCULATION")
print("="*60)

examples = [
    ("the", "city"),
    ("of", "the"),
    ("she", "was"),
    ("the", "air")
]

for w1, w2 in examples:
    p_raw = bigram_prob(w1, w2, False)
    p_smooth = bigram_prob(w1, w2, True)
    print(f"P({w2}|{w1}) -> Raw: {p_raw:.4f}, Laplace: {p_smooth:.4f}")
```

**Unigram Probability**

```python
def unigram_prob(word):
    total = sum(unigram_counts.values())
    return unigram_counts.get((word,), 0) / total

print("\nUnigram Probabilities:")
for word in ["the", "she", "city"]:
    print(f"P({word}) = {unigram_prob(word):.4f}")
```

**Trigram Probability**

```python
def trigram_prob(w1, w2, w3, smoothing=True):
    count = trigram_counts.get((w1, w2, w3), 0)
    base = bigram_counts.get((w1, w2), 0)

    if smoothing:
        return (count + 1) / (base + V)
    else:
        return count / base if base != 0 else 0

print("\nTrigram Probabilities:")
tri_examples = [
    ("the", "city", "was"),
    ("she", "was", "in"),
    ("the", "air", "was")
]

for w1, w2, w3 in tri_examples:
    p = trigram_prob(w1, w2, w3, True)
    print(f"P({w3}|{w1},{w2}) = {p:.4f}")
```

**Sentence Generation**

```python
def generate_sentence():
    word = "<s>"
    sentence = []

    for _ in range(12):
        candidates = {w2: bigram_prob(word, w2, True)
                      for (w1, w2) in bigram_counts if w1 == word}

        if not candidates:
            break

        word = random.choices(
            list(candidates.keys()),
            weights=candidates.values()
        )[0]

        if word == "</s>":
            break

        sentence.append(word)

    return " ".join(sentence)

print("\n" + "="*60)
print("STEP 6: GENERATED SENTENCES (MANUAL)")
print("="*60)

for i in range(3):
    print(f"Sentence {i+1}: {generate_sentence()}")
```

**Perplexity**

```python
train_data, vocab_nltk = padded_everygram_pipeline(2, tokenized)

model = MLE(2)
model.fit(train_data, vocab_nltk)

print("\nNLTK Generated Sentences:")
print(model.generate(7, text_seed=['forest']))
print(model.generate(7, text_seed=['leaves']))

# ------------------------------------------
# STEP 7: PERPLEXITY
# ------------------------------------------
def perplexity(sentence):
    words = ['<s>'] + sentence.lower().split() + ['</s>']
    log_prob = 0

    for i in range(1, len(words)):
        p = bigram_prob(words[i-1], words[i], True)
        log_prob += math.log2(p)

    return 2 ** (-log_prob / len(words))

test_sentences = [
    "the forest was quiet",
    "the air smelled of pine",
    "squirrel climbed tree",
    "unknown words example"
]

print("\n" + "="*60)
print("STEP 7: PERPLEXITY RESULTS")
print("="*60)

for s in test_sentences:
    print(f"{s} --> Perplexity: {perplexity(s):.2f}")
```
