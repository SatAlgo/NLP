```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

## **Text**

```python
text = """Artificial Intelligence is transforming the world.
It enables computers to understand language and make decisions."""
doc = nlp(text)
```

# **POS Tagging**

```python
print("===== POS TAGGING =====")
for token in doc:
    print(f"{token.text:15} {token.pos_:10} {token.tag_:10} {token.dep_}")
```

# **Chunking (Noun-Phrases)**

```python
print("\n===== NOUN PHRASES =====")
for chunk in doc.noun_chunks:
    print(chunk.text)
```

# **Dependency Visualization**

```python
svg = displacy.render(doc, style="dep", jupyter=False)

with open("dependency.svg", "w", encoding="utf-8") as f:
    f.write(svg)

print("Saved as dependency.svg")
```
