# **Import Libraries & Load Model**

```python
import spacy
from spacy import displacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")
```

# **Input Text**

```python
text = """
Prime Minister Narendra Modi visited Mumbai on Monday and met executives
from Reliance Industries. The meeting focused on investments worth $10 billion.
"""
```

# **Perform Named Entity Recognition**

```python
doc = nlp(text)

print("Named Entities:\n")

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

# **Visualize Entities**

```python
displacy.render(doc, style="ent", jupyter=True)
```

# **Count Frequency of Each Entity Type**

```python
labels = [ent.label_ for ent in doc.ents]
counts = Counter(labels)

print("\nEntity Frequency:\n")
for label, count in counts.items():
    print(label, ":", count)
```

# **Structured Information Extraction**

```python
info = {}

for ent in doc.ents:
    info.setdefault(ent.label_, []).append(ent.text)

print("\nStructured Information:\n")
for key, value in info.items():
    print(key, ":", value)
```
