```python
import nltk
import string

text = "Hello! My name is Shubham Sarwar and I am studying Natural Language Processing."

# Lowercase
text_lower = text.lower()

# Remove punctuation
text_clean = text_lower.translate(str.maketrans('', '', string.punctuation))

print("Original Text:", text)
print("\nLowercase Text:", text_lower)
print("\nText after removing punctuation:", text_clean)
```

# **Tokenization**

```python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize, word_tokenize

text = "Shubham Sarwar studies NLP. He practices Python programming every day."

sent_tokens = sent_tokenize(text)
word_tokens = word_tokenize(text)

print("Sentence Tokens:")
print(sent_tokens)

print("\nWord Tokens:")
print(word_tokens)
```

# **Stopwards**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')

text = "Shubham is studying NLP in the computer lab at college."

words = word_tokenize(text.lower())
stop_words = set(stopwords.words("english"))

filtered_words = list(filter(lambda word: word not in stop_words, words))

print("Original Words:")
print(words)

print("\nAfter Stopword Removal:")
print(filtered_words)
```

# **Stemming**

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

text = "Shubham is studying and learning Natural Language Processing."

words = word_tokenize(text.lower())

stemmer = PorterStemmer()

stemmed_words = list(map(stemmer.stem, words))

print("Original Words:")
print(words)

print("\nStemmed Words:")
print(stemmed_words)
```

# **Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('wordnet')

text = "Shubham was studying NLP concepts and learning new techniques."

words = word_tokenize(text.lower())

lemmatizer = WordNetLemmatizer()

lemmatized_words = list(map(lemmatizer.lemmatize, words))

print("Original Words:")
print(words)

print("\nLemmatized Words:")
print(lemmatized_words)
```

# **Word Frequency**

```python
from collections import Counter
from nltk.tokenize import word_tokenize

text = "Shubham studies NLP. Shubham practices Python for NLP assignments."

words = word_tokenize(text.lower())

freq = Counter(words)

print("Word Frequency:")
print(freq)
```
